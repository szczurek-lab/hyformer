# Analyze hparams

```python
import os
import json
import pandas as pd 

from IPython.display import HTML
```

```python
splitter = 'scaffold'
dataset = 'toxcast'
backbone = 'llama_backbone_best_big_250k_iters_final'

max_epochs = ['20']
lr_decays = ['true', 'false']
batch_sizes = ['64', '128', '256']
learning_rates = ['1e-5', '4e-5', '5e-5', '1e-4', '3e-4', '5e-4']
weight_decays = ['1e-1']
pooler_dropouts = ['0.0', '0.1', '0.2']
```

```python

df = pd.DataFrame(columns=['objective_value_mean', 'objective_value_std', 'max_epoch', 'decay_lr', 'batch_size', 'learning_rate', 'weight_decay', 'pooler_dropout'])

for max_epoch in max_epochs:
    for decay_lr in lr_decays:
        for batch_size in batch_sizes:
            for learning_rate in learning_rates:
                for weight_decay in weight_decays:
                    for pooler_dropout in pooler_dropouts:
                        
                        experiment_path = f'/lustre/groups/aih/jointformer/results_icml25_rebuttal/molecule_net/{splitter}/{dataset}/{backbone}/'\
                        f'max_epochs_{max_epoch}/decay_lr_{decay_lr}/batch_size_{batch_size}/learning_rate_{learning_rate}/weight_decay_{weight_decay}/'\
                        f'pooler_dropout_{pooler_dropout}/test_loss_aggregated.json'

                        if os.path.exists(experiment_path):
                            with open(experiment_path, 'r') as f:
                                _loss_dict = json.load(f)
                            
                            hparams = {
                                'objective_value_mean': round(_loss_dict['mean'], 3) * 100,
                                'objective_value_std': round(_loss_dict['std'], 3) * 100,
                                'max_epoch': max_epoch,
                                'decay_lr': decay_lr,
                                'batch_size': batch_size,
                                'learning_rate': learning_rate,
                                'weight_decay': weight_decay,
                                'pooler_dropout': pooler_dropout
                            }
                            hparams = {k: [v] for k,v in hparams.items()}
                            df = pd.concat([df, pd.DataFrame(hparams)], ignore_index=True)
                            
                            
                            # pd.concat([_df, ], ignore_index=True)

                        else:
                            print(f'Experiment not found: {experiment_path}')
                        

```

```python
from IPython.display import HTML

HTML(df.sort_values(by='objective_value_mean', ascending=False).to_html())
```
